```python
import matplotlib.pyplot as plt
import numpy as np
```

```python

# Random location for the word vector
word_vector = [0.2, 0.1]

# Random direction for positional encoding
positional_encoding = [0.03, 0.05]  # Adjust to avoid direct alignment

# Resultant vector
resultant_vector[0] = word_vector[0] + positional_encoding[0]
resultant_vector[1] = word_vector[1] + positional_encoding[1]

print(resultant_vector)

# Plotting
plt.figure(figsize=(8, 6))

# Plotting word vector
plt.plot(word_vector[0], word_vector[1], 'bo', label='Word Vector')
plt.text(word_vector[0], word_vector[1], '  Word Vector', verticalalignment='bottom')

# Plotting positional encoding
plt.plot(positional_encoding[0], positional_encoding[1], 'go', label='Positional Encoding')
#plt.quiver(word_vector[0], word_vector[1], positional_encoding[0], positional_encoding[1],
#           angles='xy', scale_units='xy', scale=1, color='green', label='Positional Encoding')
plt.text(positional_encoding[0], positional_encoding[1], '  Positional Encoding', verticalalignment='bottom')

# Plotting resultant vector
#plt.quiver(word_vector[0], word_vector[1], resultant_vector[0], resultant_vector[1],
#           angles='xy', scale_units='xy', scale=1, color='red', label='Resultant Vector', zorder=5) # zorder to ensure red line is on top
plt.plot(resultant_vector[0], resultant_vector[1], 'ro', label='Combined Vector')
plt.text(resultant_vector[0], resultant_vector[1], '  Combined Vector', verticalalignment='bottom')

# Axis labels and title
plt.xlabel('Dimension 1')
plt.ylabel('Dimension 2')
plt.title('Effect of Positional Encoding on Word Vector')

# Set equal aspect ratio
plt.gca().set_aspect('equal', adjustable='box')

# Adjust plot limits
x_min = min(word_vector[0], resultant_vector[0], word_vector[0] + positional_encoding[0])
x_max = max(word_vector[0], resultant_vector[0], word_vector[0] + positional_encoding[0])
y_min = min(word_vector[1], resultant_vector[1], word_vector[1] + positional_encoding[1])
y_max = max(word_vector[1], resultant_vector[1], word_vector[1] + positional_encoding[1])
plt.xlim(x_min - 0.1, x_max + 0.1)
plt.ylim(y_min - 0.1, y_max + 0.1)

ax = plt.gca()
ax.set_xlim([0.0, 0.5])
ax.set_ylim([0.0, 0.3])

# Hide grid
plt.grid(False)

# Show plot
plt.show()
```
